```python
import pandas as pd
```

```python
df = pd.read_csv("jeans.csv")
# print head
print(df.head())
#print columns 
print(df.columns)
#display info about the dataset used 
print(df.info())
```

```python
#check missing value
print(df.isnull().sum())
```

```python
#handle missing values and display results
df["seller_name"] = df["seller_name"].fillna("Unknown")

df["seller_information"] = df["seller_information"].fillna("Not Available")

df["what_customers_said"] = df["what_customers_said"].fillna("No Review")

df["videos"] = df["videos"].fillna("No Video")

df["variations"] = df["variations"].fillna("No Variations")

print(df.isnull().sum())
```

```python
#remove duplicates if any
print(df.duplicated().sum())

df = df.drop_duplicates()
```

```python
#clean final price
df["final_price"] = (
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .astype(float)
)
df["final_price"].dtype
```

```python
#create new column
df["discount_amount"] = df["initial_price"] - df["final_price"]
```

```python
#filter rows
high_rated = df[df["rating"] > 4.0]

print(high_rated[["title", "rating", "final_price"]].head())
```

```python
df.to_csv("final_cleaned_shopping_dataset.csv", index=False)
```
